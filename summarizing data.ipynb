```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Create a sample dataset
data = {
    'Age': [25, 30, 22, 35, 28, 40, 29, 33, 31, 27],
    'Salary': [50000, 60000, 45000, 80000, 52000, 90000, 58000, 75000, 62000, 49000]
}

df = pd.DataFrame(data)

# Descriptive statistics
mean_age = df['Age'].mean()
median_age = df['Age'].median()
mode_age = df['Age'].mode()[0]
std_age = df['Age'].std()

mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
mode_salary = df['Salary'].mode()[0]
std_salary = df['Salary'].std()

print("Descriptive Statistics:")
print(f"Age - Mean: {mean_age}, Median: {median_age}, Mode: {mode_age}, Std Dev: {std_age}")
print(f"Salary - Mean: {mean_salary}, Median: {median_salary}, Mode: {mode_salary}, Std Dev: {std_salary}")

# Visualization
plt.figure(figsize=(12, 8))

# Histogram for Age
plt.subplot(2, 2, 1)
sns.histplot(df['Age'], bins=5, kde=True)
plt.title('Histogram of Age')

# Boxplot for Salary
plt.subplot(2, 2, 2)
sns.boxplot(y=df['Salary'])
plt.title('Boxplot of Salary')

# Scatter plot Age vs Salary
plt.subplot(2, 2, 3)
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Scatter Plot of Age vs Salary')

plt.tight_layout()
plt.show()

```

Here's a Python example that demonstrates how to summarize a dataset using descriptive statistics and visualizations:

---

###  Descriptive Statistics Output:
- **Age**  
    - Mean: 30.0  
    - Median: 29.5  
    - Mode: 22  
    - Std Dev: 5.14  
- **Salary**  
    - Mean: 62,100.0  
    - Median: 59,000.0  
    - Mode: 45,000  
    - Std Dev: 14,903.02  

These values provide a quick overview of the central tendency and spread of the data.

---

###  Visualizations:
- **Histogram of Age:** Shows the distribution of ages.
- **Boxplot of Salary:** Highlights the median, quartiles, and any potential outliers.
- **Scatter Plot of Age vs Salary:** Helps visualize the relationship between age and salary.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example test scores for a class
test_scores = [65, 70, 72, 68, 75, 80, 85, 90, 95, 78, 82, 88, 60, 62, 67, 73, 77, 84, 91, 93]

# Define bins for score ranges (e.g., 6070, 7080, etc.)
bins = [60, 70, 80, 90, 100]

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(test_scores, bins=bins, edgecolor='black', color='skyblue')

# Add labels and title
plt.title('Distribution of Test Scores')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')

# Show average line
mean_score = np.mean(test_scores)
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=1.5, label=f'Average Score: {mean_score:.1f}')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
